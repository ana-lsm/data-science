```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./StudentPerformanceFactors.csv")

df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df = df.dropna()
df.info()
```

```python
df.duplicated().sum()
```

```python
import seaborn as sns
```

```python
sns.pairplot(data=df, hue='Distance_from_Home')
```

```python
sns.histplot(data=df['Exam_Score'])
```

```python
df.columns
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: Attendance
sns.histplot(data=df, x='Exam_Score', hue='Learning_Disabilities', kde=True, multiple="stack", ax=axes[0])
axes[0].set_title('Exam Score vs Learning Disabilities')

# Plot 2: Hours Studied
sns.histplot(data=df, x='Exam_Score', hue='Gender', kde=True, multiple="stack", ax=axes[1])
axes[1].set_title('Exam Score vs Gender')

# Plot 3: Exam Score
sns.histplot(data=df, x='Exam_Score', hue='Parental_Involvement', kde=True, multiple="stack", ax=axes[2])
axes[2].set_title('Exam Score vs Parental Involvement')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: Attendance
sns.violinplot(data=df, y='Exam_Score', x = 'Gender', ax=axes[0])
axes[0].set_title('Exam Score vs Gender')

# Plot 2: Hours Studied
sns.violinplot(data=df, y='Exam_Score', x ='Learning_Disabilities', ax=axes[1])
axes[1].set_title('Exam Score vs Learning Disabilities')

# Plot 3: Exam Score
sns.violinplot(data=df, y='Exam_Score', x = 'Parental_Involvement', split=True, inner='quart', ax=axes[2])
axes[2].set_title('Exam Score vs Parental Involvement')

# Adjust layout
plt.tight_layout()
plt.show()
```
