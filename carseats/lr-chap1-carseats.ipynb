```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/intro-stat-learning/ISLP/refs/heads/main/ISLP/data/Carseats.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df['ShelveLoc'].value_counts()
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()
```

```python
sns.pairplot(data=df, hue='US')
plt.show()
```

```python
sns.lmplot(x='Sales', y='Price', data=df, hue='US')
plt.title('Sales vs Price')
plt.show()
```

```python
sns.lmplot(x='CompPrice', y='Price', data=df, hue='US')
plt.title('Competitior Price vs Price')
plt.show()
```

```python
sns.boxplot(x='ShelveLoc', y='Sales', data=df)

plt.title('Sales por Shelve Location')
plt.show()
```

```python
df.groupby('ShelveLoc')['Sales'].mean() # Para cada tipo de localização na prateleira, qual é a média de vendas?
```

```python
sns.boxplot(x='US', y='Advertising', data=df)
plt.title('Distribuição de propaganda por região')
plt.show()
```

```python
sns.lmplot(x='Advertising', y='Sales', data=df, hue='US')
plt.title('Sales vs Advertising')
plt.show()
```
